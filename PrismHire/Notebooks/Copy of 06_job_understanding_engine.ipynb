```python
import re
import json
import pandas as pd
import numpy as np
```

```python
job_description = """
Senior Backend Engineer

Requirements:
5+ years of experience
Python
AWS
Docker
REST APIs
Microservices
Excellent communication
Ownership mindset
Fast-paced startup environment
"""
```

```python
skill_vocab = [
    "Python",
    "Java",
    "AWS",
    "Docker",
    "Kubernetes",
    "SQL",
    "React",
    "REST APIs",
    "Microservices",
    "Spark",
    "Airflow",
    "Terraform",
    "GCP",
    "Azure"
]
```

```python
def extract_skills(text):

    found = []

    text = text.lower()

    for skill in skill_vocab:
        if skill.lower() in text:
            found.append(skill)

    return found
```

```python
def extract_seniority(text):

    text = text.lower()

    if "senior" in text:
        return "senior"

    if "lead" in text:
        return "lead"

    if "manager" in text:
        return "manager"

    if "principal" in text:
        return "principal"

    return "mid"
```

```python
def get_signal(text, words):

    text = text.lower()

    for word in words:
        if word in text:
            return 1

    return 0
```

```python
def build_job_genome(text):

    genome = {}

    genome["skills"] = extract_skills(text)

    genome["seniority"] = extract_seniority(text)

    genome["leadership"] = get_signal(
        text,
        [
            "lead",
            "mentor",
            "manage",
            "team"
        ]
    )

    genome["communication"] = get_signal(
        text,
        [
            "communication",
            "stakeholder",
            "collaboration"
        ]
    )

    genome["ownership"] = get_signal(
        text,
        [
            "ownership",
            "independent",
            "self-driven"
        ]
    )

    genome["startup_adaptability"] = get_signal(
        text,
        [
            "startup",
            "fast-paced",
            "dynamic"
        ]
    )

    genome["technical_depth"] = min(
        len(genome["skills"]) / 10,
        1
    )

    return genome
```

```python
job_genome = build_job_genome(
    job_description
)

job_genome
```
