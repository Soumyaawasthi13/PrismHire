```python
import json
import pickle
from tqdm import tqdm
```

```python
candidates = []

path = "/content/drive/MyDrive/PrismHire/data/[PUB] India_runs_data_and_ai_challenge/India_runs_data_and_ai_challenge/candidates.jsonl"

with open(path, "r") as f:

    for line in tqdm(f):
        candidates.append(
            json.loads(line)
        )
```

```python
skill_set = {}

for candidate in tqdm(candidates):

    cid = candidate["candidate_id"]

    skills = set()

    for s in candidate["skills"]:

        skills.add(
            s["name"].lower()
        )

    skill_set[cid] = skills
```

```python
with open(
    "/content/drive/MyDrive/PrismHire/models/skill_set.pkl",
    "wb"
) as f:

    pickle.dump(
        skill_set,
        f
    )
```
